```python
# Defining the function
def disc_cont(df,col):
        if df[col].nunique() > 0.01*len(df):
            return('continuous')
        else:
            return('discrete')
```

```python
disc_list = []
cont_list = []
```

#### Alternative

```python
# Defining the function
def disc_cont(df,col):
        if df[col].nunique() > 0.01*len(df):
            return('continuous')
        else:
            return('discrete')
```

```python
# Using list comprehension
continous_df = numerical_df[[col for col in numerical_df.columns if disc_cont(numerical_df,col) == 'continuous']].copy()
dicrete_df = numerical_df[[col for col in numerical_df.columns if disc_cont(numerical_df,col) == 'discrete']].copy()
```
