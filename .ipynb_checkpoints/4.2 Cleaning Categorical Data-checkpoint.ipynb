# 7. Does any column contain alpha and numeric data? 
Decide how to clean it and do it now.

The only numbers in the categorical data are in the customer and in the policy_type columns. The customer values are all individual and the column can be dropped. There is no reason to change anything at policy_type, if we would remove the numbers we couldn't differentiate the policies anymore.

# 9. Check number of unique values in each column, can they be combined in any way to ease encoding? 
Comment your thoughts and make those changes.

```python
for c in categorical_df.columns:
    display(categorical_df[c].value_counts().to_frame())
```

```python
# In the column policy, we still have 6 categories.Some have few values, we simplify Corporate and Special
def pol_change(x):
    if 'Corporate' in x:
        return 'Corporate'
    elif 'Special' in x:
        return 'Special'
    else:
        return x

categorical_df['policy'] = categorical_df['policy'].apply(pol_change)
categorical_df['policy'].value_counts().to_frame()
```

```python
# The States Wshington and Nevada have very few values. We aggregate them as 'others'.
categorical_df['state'] = categorical_df['state'].apply(lambda x: 'other' if x in ['Washington', 'Nevada'] else x)
categorical_df['state'].value_counts().to_frame()
```

```python
# For education we can summarize Master and Doctor as higher education
categorical_df['education'] = categorical_df['education'].apply(lambda x: 'higher education' if x in ['Master', 'Doctor'] else x)
categorical_df['education'].value_counts().to_frame()
```

```python
# For employment_status we can summarize Medical Leave, Retired and Disabled as other        
categorical_df['employmentstatus'] = categorical_df['employmentstatus'].apply(lambda x: 'other' if x in ['Medical Leave', 'Retired', 'Disabled'] else x)
categorical_df['employmentstatus'].value_counts().to_frame()
```

```python
# For vehicle_class we can summarize Sports Car, Luxury Car and Luxury SUV as Luxury Car    
categorical_df['vehicle_class'] = categorical_df['vehicle_class'].apply(lambda x: 'Luxury Car' if x in ['Luxury SUV', 'Sports Car'] else x)
categorical_df['vehicle_class'].value_counts().to_frame()
```

# 3. There might be some columns that seem to be redundant, check their values to be sure. 
What should we do with them?

```python
# We check if we have a redundancy with crosstabs, in this case we would have overlapping values.
# We also do a contingency test
# Contingency
import scipy.stats as st
import numpy as np
```

```python
# Maybe vehicle classs and vehicle size
data_crosstab = pd.crosstab(categorical_df['vehicle_class'], categorical_df['vehicle_size'], margins=False)
display(data_crosstab)
# They don't seem to be redundant.
```

```python
st.chi2_contingency(np.array(data_crosstab))
# (12.57008448185518,                     Deviation of the values
#  0.050394366781979824,                  P-2 Value, H0: the values are independent, low means correlation
#  6,                                     Number of degrees of freedom
# array([[ 478.85485814, 3249.73381531,  892.41132654],
#         [  85.5949173 ,  580.88728229,  159.51780042],
#         [ 186.11195093, 1263.04304962,  346.84499945],
#         [ 195.43827363, 1326.33585278,  364.22587359]]))
```

```python
# H0: vehicle_class and vehicle_size are not related
# p > 0.05 ===> we can not reject the H0 ===> we can work under the assumption that they are not related 
st.chi2_contingency(data_crosstab)[1]
```

#### For other columns

```python
# Maybe coverage is the same as offer type
data_crosstab = pd.crosstab(categorical_df['coverage'], categorical_df['renew_offer_type'], margins=False)
data_crosstab
```

```python
st.chi2_contingency(data_crosstab)[1]
# They are indeed dependend on another!
```

```python
# Maybe the policy is linked to the vehicle class or size?
data_crosstab = pd.crosstab(categorical_df['policy'], categorical_df['vehicle_size'], margins=False)
data_crosstab
```

```python
st.chi2_contingency(data_crosstab)[1]
# Not dependend
```

```python
# Maybe the policy is linked to the vehicle class or size?
data_crosstab = pd.crosstab(categorical_df['policy'], categorical_df['vehicle_class'], margins=False)
data_crosstab
```

```python
st.chi2_contingency(data_crosstab)[1]
# Not dependend
```

#### Conclusion:

We found that 'coverage' and 'renew_offer_type' are indeed dependend on another, we might consider trying out our model with one of them dropped.
